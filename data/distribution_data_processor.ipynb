# distribution data processor

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = []
with open("distributions-10000000.txt", "r") as f:
    for line in f:
        data += [[int(num) for num in line.split(",")]]

titles = ["deadwood points", "number of hit Cards", "number of melds","Meld points", "number of Combos", "Combos points","number of Knock Cache cards", "Knock Cache points", "number of Load Cards","Load Card points"]
```

```python
for i in range(len(data)):
    line = data[i]
    while line[-1] == 0:
        line.pop()
    fig, axes = plt.subplots(figsize=(20,20))
    axes.bar(range(len(line)), line)
    axes.set_title(titles[i])
    if max(line) / 10000 > min(line):
        fig, axes = plt.subplots(figsize=(20,20))
        axes.set_title(titles[i] + " log")
        axes.bar(range(len(line)), line, log = True, color="orange")
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(100,25))

for i in range(5):
    line = data[i]
    while line[-1] == 0:
        line.pop()
    axes[i].bar(range(len(line)), line)
    axes[i].set_title(titles[i])
plt.tight_layout()

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(100,25))

for i in range(5):
    line = data[i + 5]
    while line[-1] == 0:
        line.pop()
    axes[i].bar(range(len(line)), line)
    axes[i].set_title(titles[i])
plt.tight_layout()
```
